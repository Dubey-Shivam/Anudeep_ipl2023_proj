Here we analyse man of the match using bar plot

````python
# Check for unique values in 'man_of_the_match', 'Home_team', 'Away_team', and 'match_no'
print(data['man_of_the_match'].unique())
print(data['Home_team'].unique())
print(data['Away_team'].unique())
print(data['match_no'].unique())
````

````python
# Prepare the data for visualization
# Create a new DataFrame that counts occurrences of Man of the Match by Team
man_of_match_counts = data.groupby(['man_of_the_match', 'winner']).size().reset_index(name='count')
````

````python
# Create a combination of match_no, team, and man of the match for the plot
# we want to include match_no as well if it provides meaningful insight in the context
man_of_match_counts['match_info'] = man_of_match_counts['man_of_the_match'] + ' - ' + man_of_match_counts['winner']
````

````python
# Create a bar plot
plt.figure(figsize=(12, 10))
sns.barplot(x='count', y='match_info', data=man_of_match_counts, palette='viridis')

# Set title and labels
plt.title('Count of Man of the Match by Team in IPL 2023')
plt.xlabel('Count')
plt.ylabel('Man of the Match - Winner Team')

# Show the plot
plt.show()
````

Here is toss winner and match result using pie chart

````python
# Data for pie chart
toss_result = data.groupby(['toss_winner', 'result']).size().reset_index(name='counts')

# Set up for pie chart
plt.figure(figsize=(10, 8))
plt.pie(toss_result['counts'], labels=toss_result['toss_winner'] + ' - ' + toss_result['result'], autopct='%1.1f%%')
plt.title('Pie Chart of Toss Winner and Match Result')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
````

````python
# Here we will count the number of wins per team
win_counts = data['winner'].value_counts().reset_index()
win_counts.columns = ['team', 'wins']

# If you want to visualize wins vs number of matches played by each team
matches_counts = data['Home_team'].value_counts().add(data['Away_team'].value_counts(), fill_value=0).reset_index()
matches_counts.columns = ['team', 'matches']

# Merge win and match count data
scatter_data = pd.merge(win_counts, matches_counts, on='team', how='outer')

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=scatter_data, x='matches', y='wins', hue='team', s=100)
plt.title('Scatter Plot of Wins vs Matches Played by Team')
plt.xlabel('Number of Matches Played')
plt.ylabel('Number of Wins')
plt.legend()
plt.show()
````

Here we analyse team and winning city using line chart

````python
# Create a DataFrame that counts wins per city for home and away teams
home_wins = data[data['winner'] == data['Home_team']].groupby(['city', 'Home_team']).size().reset_index(name='wins')
away_wins = data[data['winner'] == data['Away_team']].groupby(['city', 'Away_team']).size().reset_index(name='wins')

# Rename the columns for merging
home_wins.rename(columns={'Home_team': 'team'}, inplace=True)
away_wins.rename(columns={'Away_team': 'team'}, inplace=True)

# Combine both DataFrames
win_counts = pd.concat([home_wins, away_wins]).groupby(['city', 'team']).sum().reset_index()
print(win_counts)
````

````python
# Pivot the data to get teams as columns
pivot_win_counts = win_counts.pivot(index='city', columns='team', values='wins').fillna(0)

# Display the pivot table
print(pivot_win_counts)
````

````python
# Plotting the data
plt.figure(figsize=(12, 6))
for team in pivot_win_counts.columns:
    plt.plot(pivot_win_counts.index, pivot_win_counts[team], marker='o', label=team)

# Adding title and labels
plt.title('Number of Wins by Teams in Different Cities')
plt.xlabel('City')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)  # Rotate city names for better visibility
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')  # Legend placement
plt.grid()
plt.tight_layout()
plt.show()
````
